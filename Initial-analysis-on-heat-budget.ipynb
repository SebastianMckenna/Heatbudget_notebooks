```python
#Create fucntion to clauclate THETAO anomaly
def get_anomaly_clim(DS):
    #calc climatology
    clim = DS.groupby('time.month').mean('time')
    ANOM = DS.groupby('time.month') - clim
    return (clim, ANOM)
```

```python
def calc_advection(TA, Tbar, VA, VBAR, co_ord):
    V_DTadx = (VA*(TA.differentiate(coord = co_ord)))
    VB_DTadx = VBAR*(TA.differentiate(coord = co_ord).groupby('time.month'))
    V_dTBardx = VA.groupby('time.month')*(Tbar.differentiate(coord=co_ord))
    return (V_DTadx, VB_DTadx, V_dTBardx)
```

```python
#get zonal advection terms:
udtdx, ubardtdx, udtbardx = calc_advection(TA, Tbar, UA, Ubar, co_ord = 'lon')
```
