```python
#load in hfls data and also laod in lhf era5 data
e5_LH = xr.open_dataarray("/g/data/e14/sm2435/ERA5/clim/ERA5_lhf.nc")
outdir = '/g/data/e14/sm2435/CMIP6_HB/'#dict with file names
LHF={}
for model in os.listdir(outdir):
    #go thorugh all varis
    if model in ['NESM3']:
        pass
    else:
        mod={}
        #input file
        inp_files = (os.path.join(outdir, model, str(model+'_hfls.nc')))
        try:
            #load in file
            var = xr.open_dataarray(inp_files)
            LHF[model] = var
        except Exception as e:
            print("here comes the errorplane", e)
            
```

```python
def get_region(da, lat, lat_r, lon, lon_r):
    DA_ss = da.sel(lat=slice(lat, lat_r),lon=slice(lon, lon_r)).mean(['lat', 'lon'])
    return DA_ss
```

```python
for i, model in enumerate(LHF.keys()):
    L=LHF[model]
    L=L.assign_coords({'model':model})
    #now turn all these variables into an array
    if i == 0:
        LM = L
    else:
        LM = xr.concat([LM, L], dim = 'model')
```

```python
LM_arab = get_region(-1*LM, 10,20,50,70)
LM_setio = get_region(-1*LM, -10,0,90,110)
```

```python
LM_arab.median('model').plot(label = 'CMIP6')
plt.fill_between(range(1,13), get_region(((-1*LM.quantile(0.25, 'model'))), 10,20,50,70),
                      get_region(((-1*LM.quantile(0.75, 'model'))), 10,20,50,70), 
                      alpha = 0.25, )
get_region(e5_LH, 10,20,50,70).plot(label = 'ERA5')
plt.legend()
plt.title("Arabian Sea latent heat flux")
```

```python
LM_setio.median('model').plot(label = 'CMIP6')
plt.fill_between(range(1,13), get_region(((-1*LM.quantile(0.9, 'model'))), -10,0,90,110),
                      get_region(((-1*LM.quantile(0.1, 'model'))), -10,0,90,110), 
                      alpha = 0.25, )
get_region(e5_LH, -10,0,90,110).plot(label = 'ERA5')
plt.legend()
plt.title("SETIO latent heat flux")
```

```python
#now go through this and make a net HF term for shortwave
HF={}
for model in models.keys():
    try:
        mod = models[model]
        sw_u = mod['rsus'].rsus
        sw_d = mod['rsds'].rsds
        net_sw = sw_d-sw_u
        HF[model]=net_sw
    except Exception as e:
        print("Oh no, here is your model and error!", model, e)
```
