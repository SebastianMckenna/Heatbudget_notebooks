```python
#load in SST

A_SST = xr.open_dataset('/home/z5113258/Documents/data/CMIP6/ACCESS-CM2/TOS/tos_Omon_ACCESS-CM2_historical_r1i1p1f1_gn_185001-201412.nc.nc').tos
A_SST = A_SST.sel(lat=slice(-40,30),lon=slice(30,125))
A_SST = A_SST.sel(time= slice('1950', '2015'))


#load in SST
G_SST = xr.open_dataset('/home/z5113258/Documents/GODAS/remapcon_1/sst_con.nc').sst
LSM = xr.open_dataset('/home/z5113258/Documents/GODAS/remapcon_1/lsmask_con.nc').mask
G_SST= G_SST.where(LSM[0])
G_SST = G_SST.sel(lat=slice(-40,30),lon=slice(30,125))
```

```python
#get SSTA
#us ref period 1960-1990
ASST_ref = A_SST.sel(time= slice('1950', '1990')).groupby('time.month').mean('time')
GSST_ref = G_SST.sel(time= slice('1980', '2000')).groupby('time.month').mean('time')
```

```python
A_SSTA = A_SST.groupby('time.month') - ASST_ref
G_SSTA = G_SST.groupby('time.month') - GSST_ref
```

```python
#get region
def get_region(DS, lat, lat_r, lon, lon_r):
    setio = DS.sel(lat=slice(lat, lat+lat_r),lon=slice(lon, lon+lon_r)).mean(['lat', 'lon'])
    return setio
A_SCT = get_region(A_SSTA, -12,7,50,20)
G_SCT = get_region(G_SSTA, -12,7,50,20)
```

```python
diff = (A_SSTA.std(dim='time')-G_SSTA.std(dim='time'))
```

```python
def plot_background(ax):
    #for use with plotting 12 months. need to change 
    ax.set_extent([30., 125., -20., 20.])
    ax.add_feature(cfeature.COASTLINE.with_scale('50m'), linewidth=0.5)
    ax.add_feature(cfeature.LAND)
    gl = ax.gridlines(draw_labels=True, dms=True, x_inline=False, y_inline=False)
    gl.top_labels = False
    gl.right_labels = False

    return ax
import matplotlib.patches as patches
from shapely import geometry
```

```python
fig, axes = plt.subplots(nrows=1, ncols=1, subplot_kw={'projection':ccrs.PlateCarree()})


d = axes.pcolormesh(diff.lon, diff.lat, diff, 
                                cmap = 'bwr', 
                                transform=ccrs.PlateCarree())
axes.set_title('ACCESS-CM2 - OISSTv2 SSTA \n standard deviation difference')
axes.add_patch(patches.Rectangle((50,-12), 20, 7, color="k", fill = False))
plot_background(axes)
fig.colorbar(d, ax = axes)
```
