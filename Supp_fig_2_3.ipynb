```python
import xarray as xr
import dask
import matplotlib.pyplot as plt
import numpy as np
import os


import matplotlib as mpl
import scipy
import glob as glob
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import calendar
import xskillscore as xs
from scipy.stats import iqr
```

```python
datadir = '/home/z5113258/Documents/data/CMIP6_HB_new/'
#dict with file names
varis = {'tx':'_tauu.nc', 'ty':'_tauv.nc'}
models={}
#get all data for temp tendency
for model in os.listdir(datadir):
    #go thorugh all varis
    if model in ['NESM3']:
        pass
    else:
        mod={}
        for key in varis.keys():
            #input file
            inp_files = (os.path.join(datadir, model, str(model+varis[key])))
            try:
                #load in file
                var = xr.open_dataarray(inp_files)
                #put into dict
                mod[key]=var
            except Exception as e:
                print(e)
                pass
        #now append data to model dict
        models[model] = mod
del models["not_full"]
```

```python
datadir = '/home/z5113258/Documents/data/REAN_HB/'
#dict with file names
varis = {'tx':'_tauu.nc', 'ty':'_tauv.nc'}
obs={}
#get all data for temp tendency
for model in os.listdir(datadir):
    #go thorugh all varis
    mod={}
    for key in varis.keys():
        #input file
        inp_files = (os.path.join(datadir, model, str(model+varis[key])))
        try:
            #load in file
            var = xr.open_dataarray(inp_files)
            #put into dict
            mod[key]=var
            #now append data to model dict
            obs[model] = mod
        except Exception as e:
            print(e)
            pass
```

```python
for i, mod in enumerate(models.keys()):
    tauu = models[mod]['tx']
    tauu = tauu.assign_coords({"model":i})
    if i == 0:
        tauum = tauu
    else:
        tauum = xr.concat([tauum, tauu], dim = 'model')
```

```python
for i, mod in enumerate(models.keys()):
    tauv = models[mod]['ty']
    tauv = tauv.assign_coords({"model":i})
    if i == 0:
        tauvm = tauv
    else:
        tauvm = xr.concat([tauvm, tauv], dim = 'model')
```

```python
tauuMM = tauum.median('model')
tauvMM = tauvm.median('model')
```

```python
#era5 wind stress
E5x = obs['ERA5']['tx']
E5y = obs['ERA5']['ty']
```

```python
XB = tauuMM - E5x
YB = tauvMM - E5y
```

```python
spdT = xr.ufuncs.sqrt(XB**2 +YB**2)
```

```python
spd = xr.ufuncs.sqrt(tauuMM**2 +tauvMM**2)
```

```python
spdE5 = xr.ufuncs.sqrt(E5x**2 +E5y**2)
```

```python
spdB = spd - spdE5
```

```python
#make the wind bias quivers colors smaller
XBM = XB.sel(lat=slice(-30, 30),lon=slice(30, 130))
YBM = YB.sel(lat=slice(-30, 30),lon=slice(30, 130))
spdB = spdB.sel(lat=slice(-30, 30),lon=slice(30, 130))
```

```python
import matplotlib.patches as patches
```

```python
plt.rcParams['font.size'] = 14
fg = spdB.plot.contourf(col='month', col_wrap = 4,  transform=ccrs.PlateCarree(),
                             subplot_kws={"projection": ccrs.PlateCarree()}, 
                              figsize=(16,8), vmin = -0.12, vmax = 0.12, levels =13,  cmap = 'RdBu_r',#can do 11 or 21 levels
                                cbar_kwargs=dict(label='speed bias N*m^-2'), extend = 'both') 
#Add in features
for i, ax in enumerate(fg.axes.flat):
    if i ==0:
        Q = ax.quiver(XBM.lon[::3], XBM.lat[::3], XBM[i, ::3,::3], YBM[i, ::3,::3], 
                  transform=ccrs.PlateCarree(), scale = 1)
    else:
        ax.quiver(XBM.lon[::3], XBM.lat[::3], XBM[i, ::3,::3], YBM[i, ::3,::3], 
                  transform=ccrs.PlateCarree(), scale = 1)
    #ax.coastlines()
    ax.set_facecolor('grey')
    gl = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True,
                      linewidth=0.7, color='gray', alpha=0.5, linestyle='--')
    if i in [0,1,2,3,4,5,6,7]:
        gl.bottom_labels = False
    if i in [1,2,3,5,6,7,9,10,11]:
        gl.left_labels = False
    gl.top_labels = False
    gl.right_labels = False
    #add in the boxes of interest
    if i == 0:
        ax.add_patch(patches.Rectangle((50,-10), 20, 20, color="k", fill = False, linewidth=1))
        ax.add_patch(patches.Rectangle((50,10), 20, 10, color="k", fill = False, linewidth=1))
        ax.add_patch(patches.Rectangle((90,-10), 20, 10, color="k", fill = False, linewidth=1))
#now add the month labels
titles =['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun','Jul', 
         'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
for ax,title in zip(fg.axes.flatten(),titles):
    ax.set_title(title )
qk = ax.quiverkey(Q, 1, 1.1, 0.05, r'$0.05 \frac{N}{m^2}$', labelpos='E',
                   coordinates='axes')
#plt.savefig('/home/z5113258/Documents/figures/Paper_23-01-30/SF_3.eps', pad_inches = 1,  bbox_inches = 'tight',)
#plt.savefig('/home/z5113258/Documents/figures/Figures_high_res/SF_N3.png',pad_inches = 1, dpi=600, bbox_inches = 'tight', facecolor = 'white')
```
